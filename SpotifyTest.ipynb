```python
#Import and authenticate

import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials())
```

```python
# serach for artist 
name = "The Beatles"

results = sp.search(q='artist:' + name, type='artist')
```

```python
#print artist ID
items = results['artists']['items']
artist = items[0]
artist_id = artist['id']
print (f"artist id =  {artist_id}")
print (f"artist name =  {artist['name']}")
```

```python
# Get top 10 tracks

artist_top_tracks = sp.artist_top_tracks(artist_id)

for track in artist_top_tracks['tracks'][:10]:
    print('track    : ' + track['name'])
    print(f"Track Id : {track['id']}")
    print()
```

```python
#Get more information about Track
trackid = "6dGnYIeXmHdcikdzNNDMm2"
track_data = sp.track(trackid)
print(f" release date of the album {track_data['album']['release_date']}")
print(f" Name of the album - {track_data['album']['name']}")
print(f"Duration of the track - {track_data['duration_ms']}")
```

```python
# Get audio feature of the track

track_audio_feature = sp.audio_features(trackid)
track_audio_feature
```

```python
# get audio analysis of track
track_audio_analysis = sp.audio_analysis(trackid)
track_audio_analysis
```
